```julia
using Plots
```

Run the following block of code to randomly choose sine wave parameters.

```julia
A = rand(1:20) / 2
f = rand(1:100) * 10
nCycles = rand(2:10) / 2
T = 1 / f
duration = nCycles * T

println("Please plot $(A)sin(2Ï€$(f)t) for $(round(duration, sigdigits=3)) seconds")
```

Once you've drawn your own plot, run the following code to plot the sine wave.

```julia
println("Plotting for $(nCycles) cycles")
T = 1 / f
s(x) = A*sin(2Ï€*f*x)
x = [x for x in range(0, duration, step=0.000001)]
y = s.(x)
xAxisTicks = round.([n * T for n in range(0, nCycles, step=0.25)], sigdigits=3)
plot(x, y,
    xticks=xAxisTicks,
    yticks=[-A, 0, A],
    xlabel="Time (s)",
    ylabel="Amplitude",
    legend=false)
```
